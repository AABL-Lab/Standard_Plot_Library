In this notebook, we will walk through how to use matplotlib to make legible and exciting pie charts! Graphs are an integeral part of every paper, so make sure yours stand out!

We have 4 key imports:


*   Numpy will allow us to easily work with array data structures and preform useful operations on our data.
*   matplotlib.pyplot allows us to graph our data and the "style" import allows us to costomize the colors and overall look of our graphs.
* scipy stats is a common, but powerful, stats library that will take care of all of our statistical needs!






```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from scipy import stats
```


First let's generate some mock data. We will have data respresenting participant aggreement with a Likert Style assessment of robot likability. Consider this data that was gathered during a survey of participants perceptions of robot likability. 

We can create correspondings angles and labels for our mock data.


```python
# Mock data:
labels = 'Strongly Agree', 'Somewhat Agree', 'Neither Agree nor Disagree', 'Somewhat Disagree', 'Strongly Disagree'
sizes = [15, 30, 35, 10, 10]

print(f"Shape of label array: {np.shape(labels)}")
print(f"Shape of size array: {np.shape(sizes)}")
```


Shape of robot array: (3, 99)
Shape of human array: (3, 99)

```python

fig1, ax1 = plt.subplots()
ax1.pie(sizes)
plt.show()
```

Putting the data directly into a pie chart produces a circular shape with no labels or axis. We can customize this to make it more readible, by specifying the parameters of the pie chart. In the example below, we have added a shadow to the graph ('shadow'), percent labels to each plot('autopct'), and labels for what each chuck indicates 

```python

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=False)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

You can customize the rotation of the chart using the start angle function.

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

You can also add a legend to clarify the data, however on it's own it will appear directly on top of the pie chart. Use the bbox_to_anchor parameter to customize the legend so it appears in a legible location.

Similarly, a title can be added to the graph using plt.title.

```python
fig1, ax1 = plt.subplots()
explode = (0, 0.1, 0,0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# To see the labels, we need a legend! Make sure your legends are big 
# and easy to read! 
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, fontsize="xx-large")
plt.title("User response to statement 'the robot is likable'")
plt.show()
```

We might also want to create custom colors for our graph slices. We can do this using an array of specified color values. Let's also make sure our graph is color-blind friendly by selecting a color blind friendly palette.

You can check your palette at https://www.color-blindness.com/coblis-color-blindness-simulator/ to make sure it is legible. 

```python
fig1, ax1 = plt.subplots()
color_set = ('#F3CD3E','#BA21EF','#7DA0EA','#4AFEF9','#FF4300')
explode = (0, 0.1, 0,0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
ax1.pie(sizes, explode=explode, colors=color_set, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# To see the labels, we need a legend! Make sure your legends are big 
# and easy to read!


plt.legend(bbox_to_anchor=(1.1, 1), loc=2, fontsize="xx-large")
plt.title("User response to statement 'the robot is likable'")
plt.show()
```
