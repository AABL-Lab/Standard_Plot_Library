Welcome! In this notebook we will walk through how to use matplotlib to make legible and exciting looking dot plots! Graphs are a an integeral part of every paper, so make sure yours stand out!

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from scipy import stats


# Mock data:
robot_data = np.array([[np.log(i * np.random.uniform(.2, 1)) for i in range(1, 100)],
              [np.log(i * np.random.uniform(.2, 1)) for i in range(1, 100)],
              [np.log(i * np.random.uniform(.2, 1)) for i in range(1, 100)]])
human_data = np.array([[np.log(3*i*np.random.uniform(.2, 1)) for i in range(1, 100)],
              [np.log(3*i*np.random.uniform(.2, 1)) for i in range(1, 100)],
              [np.log(3*i*np.random.uniform(.2, 1)) for i in range(1, 100)]])


# Calculate averages across 3 trials
robot_data_avg = np.mean(robot_data, 0)
human_data_avg = np.mean(human_data, 0)


#Creates of x values for the data 
xarr = []
for i in range(0, 99):
  xarr.append(i)

#Seperating the figue and axis so they are easier to work with
fig, ax = plt.subplots() 


#Remove the lines on the top and right side of the graph
right_side = ax.spines["right"]
top_side = ax.spines["top"]
right_side.set_visible(False)
top_side.set_visible(False)

#Set the range for the tick marks on the y axis
yticks = [i for i in range(7)]

#Plot the data on the scatter plot
#Note: this is where the color of each data set can be changed
plt.scatter(xarr, robot_data_avg, label='Label 1', color='red', s=25, marker="o")
plt.scatter(xarr, human_data_avg, label='Label 2', color='blue', s=25, marker="o")

#Make the x and y ticks legible
plt.yticks(yticks, fontsize=16) 
plt.xticks(fontsize=16)

# To see the labels, we need a legend! Make sure your legends are big 
# and easy to read! 
plt.legend(fontsize="xx-large")

#Plots the x axis, y axis, and title labels
plt.xlabel("X Label", fontsize="xx-large")
plt.ylabel("Y Label", fontsize="xx-large")
plt.title("Title", fontsize="xx-large")

plt.show()

```
